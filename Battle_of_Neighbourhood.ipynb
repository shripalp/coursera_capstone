```python
#import required libraries
import pandas as pd
import numpy as np
```

```python
#import Calgary neighbourhood coordinates
url ='https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_T'
```

```python
#rscrape website and load to Pandas dataframe
dfs = pd.read_html(url)

#get the table into data-frame
df_calgary = dfs[1]
df_calgary.head(16)

```

```python
df_calgary.rename(columns = {'Postal Code': 'PostalCode'}, inplace = True)
df_calgary.head()
```

```python
df_calgary.shape
```

```python
indexNames = df_calgary[ (df_calgary['Borough'] != 'Calgary')].index
df_calgary.drop(indexNames, inplace = True)

df_calgary = df_calgary.reset_index(drop=True)

df_calgary.head()
```

```python
df_calgary.shape
```

```python
#drop rows of data where Borough and Neighbourhood not assigned to postal code
indexNames = df_calgary[(df_calgary['Latitude'] == 'Not assigned')].index
df_calgary.drop(indexNames, inplace = True)

df_calgary = df_calgary.reset_index(drop=True)

df_calgary.head(40)
```
